pandas allows developer to execute function on data frames

```python
def short_parent_education(parent_education):
    if parent_education == "Bachelor":
        return "B"
    elif parent_education == "High School":
        return "HS"
    else:
        return "HQ"

df["parent_education_short"] = df["parent_education"].apply(short_parent_education)
df[["student_id", "parent_education", "parent_education_short"]]
```

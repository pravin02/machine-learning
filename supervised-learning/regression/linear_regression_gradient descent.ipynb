```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
X = np.array([500, 750, 1000, 1250, 1500, 1750, 2000]).reshape(-1, 1)
Y = np.array([150, 200, 250, 300, 350, 400, 450])
print(f"X= {X}")
print(f"Y= {Y}")
```

```python
model = LinearRegression()
model.fit(X, Y)
```

```python
size_to_predict = np.array([[1600]])
predicted_price = model.predict(size_to_predict)
print(f"Predicted price for 1600 sq ft: ${predicted_price[0]*1000:.2f}")
```

```python
plt.scatter(X, Y, color="blue", label="Actual Data")
plt.plot(X, model.predict(np.array([[1300], [1500], [1000], [1600], [350], [500], [450]])), color="red", label="Regression Line")
plt.xlabel("House Size (Sq. Ft.)")
plt.ylabel("Price ($1000s)")
plt.legend()
plt.show()
```
